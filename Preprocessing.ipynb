```python
import pandas as pd
import numpy as np

origin_data = pd.read_csv('origin_data.csv')
origin_data = origin_data.drop(['humidity'], axis=1)
usable_data = origin_data[origin_data['STAT_CAUSE_CODE'] == 9]
usable_data = usable_data.drop(['STATE', 'STAT_CAUSE_CODE', 'STAT_CAUSE_DESCR', 'CONT_DOY', 'FIRE_SIZE_CLASS', 'LATITUDE', 'LONGITUDE'], axis=1)

humidity_data = pd.read_csv('humidity.csv', index_col='datetime')
pressure_data = pd.read_csv('pressure.csv', index_col='datetime')
temperature_data = pd.read_csv('temperature.csv', index_col='datetime')
wind_direction_data = pd.read_csv('wind_direction.csv', index_col='datetime')
wind_speed_data = pd.read_csv('wind_speed.csv', index_col='datetime')

humidities = []
pressures = []
temperatures = []
wind_directions = []
wind_speeds = []

times = len(usable_data.values)
for i in range(times):
    try:
        humidity = humidity_data.loc[usable_data.values[i][1], usable_data.values[i][0]]
    except:
        humidity = None
        
    try:
        pressure = pressure_data.loc[usable_data.values[i][1], usable_data.values[i][0]]
    except:
        pressure = None
        
    try:
        temperature = temperature_data.loc[usable_data.values[i][1], usable_data.values[i][0]]
    except:
        temperature = None
        
    try:
        wind_direction = wind_direction_data.loc[usable_data.values[i][1], usable_data.values[i][0]]
    except:
        wind_direction = None
        
    try:
        wind_speed = wind_speed_data.loc[usable_data.values[i][1], usable_data.values[i][0]]
    except:
        wind_speed = None
    
    humidities.append(humidity)
    pressures.append(pressure)
    temperatures.append(temperature)
    wind_directions.append(wind_direction)
    wind_speeds.append(wind_speed)

usable_data['Humidity'] = humidities
usable_data['Pressure'] = pressures
usable_data['Temperature'] = temperatures
usable_data['Wind_direction'] = wind_directions
usable_data['Wind_speed'] = wind_speeds
usable_data = usable_data.dropna(axis=0,how='any')

usable_data.to_csv('usable_data.csv')
```
